````python
lm_url_instruments = "https://paper.lemon.markets/rest/v1/instruments/"
lm_url_token = "https://auth.lemon.markets/oauth2/token"
````

````python
session = aiohttp.ClientSession()
````

````python
token_body = {
    "client_id": 'XXX',
    "client_secret": 'XXX',
    "grant_type": "client_credentials"
}
````

Get token and create bearer token header for requests.

````python
response = await session.post(lm_url_token, data=token_body)
body = await response.text()
content = json.loads(body)
token = content['access_token']
header = {"Authorization": f"Bearer {token}"}
````

Function that returns a tuple if the first 100 results contained the given ticker symbol or not.

````python
async def ticker_exists_in_lm(ticker='ABB'):
    response = await session.get(lm_url_instruments + f"?search={ticker}", headers=header)
    body = await response.text()
    try:
        content = json.loads(body)
    except:
        print(body)
        return (ticker, False)
        
    exists = bool(len(list(filter(lambda x: x['symbol'] == ticker, content["results"]))))
    return (ticker, exists)
````

Example for one call.

````python
await ticker_exists_in_lm()
````

Percentage of tickers found.

````python
tasks = [ticker_exists_in_lm(ticker) for ticker in ticker_list]
results = await asyncio.gather(*tasks)
np.mean([res[1] for res in results]) * 100
````
